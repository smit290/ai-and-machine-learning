```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.decomposition import PCA

# Step 1: Load image and convert to grayscale
image = Image.open("/content/lena_gray.gif").convert("L")
image_array = np.array(image)
height, width = image_array.shape

# Step 2: Reshape the image (optional, PCA can also work on 2D)
# Here, each row of the image is an observation
data = image_array

# Step 3: Apply PCA
# Choose the number of principal components to keep
k = 50  # You can change this value for more/less compression
pca = PCA(n_components=k)
pca.fit(data)

# Transform the data into principal component space
transformed_data = pca.transform(data)

# Reconstruct the image from top k components
reconstructed_data = pca.inverse_transform(transformed_data)

# Step 4: Display original and reconstructed images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image_array, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Reconstructed image
plt.subplot(1, 2, 2)
plt.imshow(reconstructed_data, cmap='gray')
plt.title(f"Reconstructed Image\nwith {k} PCs")
plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Compressed Representation (PCA-transformed)
plt.subplot(1, 3, 2)
plt.imshow(transformed_data, cmap="gray", aspect="auto")
plt.title(f"Compressed Image (k={k})")
plt.axis("off")

# Reconstructed / Decompressed Image
plt.subplot(1, 3, 3)
plt.imshow(reconstructed_data, cmap="gray")
plt.title("Decompressed Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
