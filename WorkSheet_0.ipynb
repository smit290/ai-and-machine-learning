Task-1

Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

```python
def convert_units(category, from_unit, to_unit, value):
    """
    Convert between different units of measurement.

    Parameters:
    category (str): The measurement category - 'length', 'weight', or 'volume'
    from_unit (str): The unit to convert from
    to_unit (str): The unit to convert to
    value (float): The numerical value to convert

    Returns:
    float: The converted value
    """
    # Length conversions
    if category == "length":
        if from_unit == "m" and to_unit == "ft":
            return value * 3.281
        elif from_unit == "ft" and to_unit == "m":
            return value / 3.281

    # Weight conversions
    elif category == "weight":
        if from_unit == "kg" and to_unit == "lbs":
            return value * 2.205
        elif from_unit == "lbs" and to_unit == "kg":
            return value / 2.205

    # Volume conversions
    elif category == "volume":
        if from_unit == "L" and to_unit == "gal":
            return value * 0.264172
        elif from_unit == "gal" and to_unit == "L":
            return value / 0.264172

    return None

def get_numeric_input(prompt):
    """
    Get numeric input from user with error handling.

    Parameters:
    prompt (str): The prompt to display to user

    Returns:
    float: The numeric value entered by user
    """
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Invalid input! Please enter a number.")

def main():
    """Main program function to run the unit converter."""

    measure = ["Length", "Weight", "Volume"]
    print("📏 Welcome to Unit Converter!")
    print("Please type 1 for Length, 2 for Weight and 3 for Volume")

    try:
        choice = input("Enter your choice (1, 2, or 3): ")

        if choice == "1":  # Length
            print("\n Length Conversion")
            print("Enter 1 to convert meter to feet, or enter 2 to convert feet to meter")
            lg = input("Enter your choice (1, 2): ")

            if lg == "1":
                value = get_numeric_input("Enter the value in meters: ")
                result = convert_units("length", "m", "ft", value)
                print(f"{value} meters = {result:.2f} feet")

            elif lg == "2":
                value = get_numeric_input("Enter the value in feet: ")
                result = convert_units("length", "ft", "m", value)
                print(f"{value} feet = {result:.2f} meters")

            else:
                print(" Please enter 1 or 2")

        elif choice == "2":  # Weight
            print("\nWeight Conversion")
            print("Enter 1 to convert kg to pounds, or enter 2 to convert pounds to kg")
            kg = input("Enter your choice (1, 2): ")

            if kg == "1":
                value = get_numeric_input("Enter the value in kilograms: ")
                result = convert_units("weight", "kg", "lbs", value)
                print(f"{value} kg = {result:.2f} lbs")

            elif kg == "2":
                value = get_numeric_input("Enter the value in pounds: ")
                result = convert_units("weight", "lbs", "kg", value)
                print(f" {value} lbs = {result:.2f} kg")

            else:
                print(" Please enter 1 or 2")

        elif choice == "3":  # Volume
            print("\n Volume Conversion")
            print("Enter 1 to convert liters to gallons, or enter 2 to convert gallons to liters")
            vol = input("Enter your choice (1, 2): ")

            if vol == "1":
                value = get_numeric_input("Enter the value in liters: ")
                result = convert_units("volume", "L", "gal", value)
                print(f" {value} L = {result:.2f} gal")

            elif vol == "2":
                value = get_numeric_input("Enter the value in gallons: ")
                result = convert_units("volume", "gal", "L", value)
                print(f"{value} gal = {result:.2f} L")

            else:
                print("Please enter 1 or 2")

        else:
            print("Invalid choice! Please enter 1, 2, or 3.")

    except Exception as e:
        print(f"An error occurred: {e}")

# to run the program
if __name__ == "__main__":
    main()
```

```python
def calculate_sum(numbers):
    """
    Calculate the sum of all numbers in a list.

    Parameters:
    numbers (list): A list of numeric values

    Returns:
    float: The sum of all numbers
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculate the average of numbers in a list.

    Parameters:
    numbers (list): A list of numeric values

    Returns:
    float: The average of all numbers

    Raises:
    ValueError: If the list is empty
    """
    if len(numbers) == 0:
        raise ValueError("Cannot calculate average of an empty list")
    return sum(numbers) / len(numbers)

def calculate_maximum(numbers):
    """
    Find the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values

    Returns:
    float: The maximum value in the list

    Raises:
    ValueError: If the list is empty
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find maximum of an empty list")
    return max(numbers)

def calculate_minimum(numbers):
    """
    Find the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values

    Returns:
    float: The minimum value in the list

    Raises:
    ValueError: If the list is empty
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find minimum of an empty list")
    return min(numbers)

def get_numbers_from_user():
    """
    Prompt user to enter numbers and convert them to a list of floats.

    Returns:
    list: A list of float numbers entered by the user

    Raises:
    ValueError: If no numbers are entered or input contains non-numeric values
    """
    user_input = input("Enter a list of numbers separated by spaces: ").strip()

    # Check if input is empty
    if not user_input:
        raise ValueError("No numbers entered. Please enter at least one number.")

    # Split the input and convert to float
    try:
        numbers = [float(num) for num in user_input.split()]
    except ValueError:
        raise ValueError("Invalid input! Please enter only numbers separated by spaces.")

    return numbers

def display_menu():
    """Display the operation menu to the user."""
    print("\n" + "="*40)
    print("MATHEMATICAL OPERATIONS MENU")
    print("="*40)
    print("1. Calculate SUM of numbers")
    print("2. Calculate AVERAGE of numbers")
    print("3. Find MAXIMUM value")
    print("4. Find MINIMUM value")
    print("5. Exit")
    print("="*40)

def main():
    """
    Main program function to run the mathematical operations tool.
    """
    print("Welcome to the Mathematical Operations Program!")

    while True:
        # Display menu and get user choice
        display_menu()
        choice = input("Enter your choice (1-5): ").strip()

        # Exit condition
        if choice == "5":
            print("Thank you for using the program. Goodbye!")
            break

        # Validate choice
        if choice not in ["1", "2", "3", "4"]:
            print("Invalid choice! Please enter a number between 1 and 5.")
            continue

        # Get numbers from user
        try:
            numbers = get_numbers_from_user()
            print(f"\nYou entered: {numbers}")
        except ValueError as e:
            print(f"Error: {e}")
            continue

        # Perform selected operation
        if choice == "1":
            result = calculate_sum(numbers)
            print(f"SUM of {numbers} = {result}")

        elif choice == "2":
            try:
                result = calculate_average(numbers)
                print(f"AVERAGE of {numbers} = {result:.2f}")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "3":
            try:
                result = calculate_maximum(numbers)
                print(f"MAXIMUM value in {numbers} = {result}")
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "4":
            try:
                result = calculate_minimum(numbers)
                print(f"MINIMUM value in {numbers} = {result}")
            except ValueError as e:
                print(f"Error: {e}")

        # Ask if user wants to continue
        again = input("\nDo you want to perform another operation? (y/n): ").strip().lower()
        if again != 'y':
            print("Thank you for using the program. Goodbye!")
            break

# Run the program
if __name__ == "__main__":
    main()
```

```python
def extract_every_other(lst):
    """
    Extracts every other element from a list, starting from the first element.

    Args:
        lst (list): The input list.

    Returns:
        list: A new list containing every other element.
    """
    return lst[::2]

# Example usage:
input_list = [1, 2, 3, 4, 5, 6]
output_list = extract_every_other(input_list)
print(f"Original list: {input_list}")
print(f"Every other element: {output_list}")

input_list_2 = ['a', 'b', 'c', 'd', 'e']
output_list_2 = extract_every_other(input_list_2)
print(f"\nOriginal list: {input_list_2}")
print(f"Every other element: {output_list_2}")
```

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from a given list starting at index 'start'
    and ending at index 'end' (inclusive).

    Args:
        lst (list): The input list
        start (int): Starting index
        end (int): Ending index

    Returns:
        list: Sublist from start to end
    """
    return lst[start:end+1]


# Example usage
input_list = [1, 2, 3, 4, 5, 6]
result = get_sublist(input_list, 2, 4)

print("Original list:", input_list)
print("Sublist from index 2 to 4:", result)
```

```python
def reverse_list(lst):
    """
    Reverses a list using slicing.

    Args:
        lst (list): The input list

    Returns:
        list: Reversed list
    """
    return lst[::-1]


# Example usage
input_list = [1, 2, 3, 4, 5]
reversed_list = reverse_list(input_list)

print("Original list:", input_list)
print("Reversed list:", reversed_list)
```

```python
def remove_first_last(lst):
    """
    Removes the first and last elements of a list using slicing.

    Args:
        lst (list): The input list

    Returns:
        list: A list without the first and last elements
    """
    return lst[1:-1]


# Example usage
input_list = [1, 2, 3, 4, 5]
result = remove_first_last(input_list)

print("Original list:", input_list)
print("List without first and last elements:", result)
```

```python
def get_first_n(lst, n):
    """
    Returns the first n elements from a list using slicing.

    Args:
        lst (list): The input list
        n (int): Number of elements to extract

    Returns:
        list: First n elements of the list
    """
    return lst[:n]


# Example usage
input_list = [1, 2, 3, 4, 5]
result = get_first_n(input_list, 3)

print("Original list:", input_list)
print("First 3 elements:", result)
```

```python
def get_last_n(lst, n):
    """
    Returns the last n elements from a list using slicing.

    Args:
        lst (list): The input list
        n (int): Number of elements to extract from the end

    Returns:
        list: Last n elements of the list
    """
    return lst[-n:]


# Example usage
input_list = [1, 2, 3, 4, 5]
result = get_last_n(input_list, 2)

print("Original list:", input_list)
print("Last 2 elements:", result)
```

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Args:
        lst (list): A nested list

    Returns:
        list: A flattened list
    """
    flat_list = []
    for sublist in lst:
        flat_list.extend(sublist)
    return flat_list


# Example usage
input_list = [[1, 2], [3, 4], [5]]
result = flatten(input_list)

print("Original nested list:", input_list)
print("Flattened list:", result)
```

```python
def access_nested_element(lst, indices):
    """
    Returns a specific element from a nested list using given indices.

    Args:
        lst (list): The nested list
        indices (list): List of indices to access the element

    Returns:
        element: The element at the specified position
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# Example usage
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

result = access_nested_element(lst, indices)

print("Nested list:", lst)
print("Element at indices", indices, ":", result)
```

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list.

    Args:
        lst (list): A nested list containing numbers

    Returns:
        int/float: Sum of all elements
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # recursive call
        else:
            total += item
    return total


# Example usage
input_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(input_list)

print("Nested list:", input_list)
print("Sum of all elements:", result)
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Args:
        lst (list): Nested list
        elem: Element to remove

    Returns:
        list: Modified nested list
    """
    result = []

    for sublist in lst:
        new_sublist = []
        for item in sublist:
            if item != elem:
                new_sublist.append(item)
        result.append(new_sublist)

    return result


# Example usage
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

result = remove_element(lst, elem)

print("Original list:", lst)
print("List after removing element:", result)
```

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list of any depth.

    Args:
        lst (list): A nested list containing numbers

    Returns:
        int/float: Maximum element
    """
    maximum = float('-inf')  # start with negative infinity

    for item in lst:
        if isinstance(item, list):
            # Recursively find the max in the sublist
            sub_max = find_max(item)
            if sub_max > maximum:
                maximum = sub_max
        else:
            if item > maximum:
                maximum = item

    return maximum


# Example usage
nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)

print("Nested list:", nested_list)
print("Maximum element:", result)
```

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Args:
        lst (list): Nested list
        elem: Element to count

    Returns:
        int: Number of occurrences
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            # Recursively count in the sublist
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


# Example usage
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

result = count_occurrences(lst, elem)

print("Nested list:", lst)
print(f"Occurrences of {elem}:", result)
```

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Args:
        lst (list): Deeply nested list

    Returns:
        list: Flattened list
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            # Recursively flatten the sublist
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list


# Example usage
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", result)
```

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Args:
        lst (list): Nested list

    Returns:
        float: Average of all elements
    """
    def flatten(lst):
        """Helper function to flatten nested lists"""
        flat_list = []
        for item in lst:
            if isinstance(item, list):
                flat_list.extend(flatten(item))
            else:
                flat_list.append(item)
        return flat_list

    flat_list = flatten(lst)
    total = sum(flat_list)
    count = len(flat_list)
    return total / count if count != 0 else 0


# Example usage
nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print("Nested list:", nested_list)
print("Average of all elements:", result)
```
